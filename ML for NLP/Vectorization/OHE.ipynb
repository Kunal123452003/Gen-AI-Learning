#### Creating the vocabulary from the text data

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
```

```python
my_vocab = set()
for sentence in sentiment_data['text']:
    words = tokenizer.tokenize(sentence.lower())
    # print(f"Tokenized words: {words}")
 
    my_vocab.update(words)
    # print(f"Current vocabulary size: {len(my_vocab)}")
    # print(f"Current vocabulary: {my_vocab}")
```

```python
len(my_vocab)     # our vocabulary size
```

```python
print(my_vocab)
```

#### Tokenizing the vocabulary

```python
token = [ [word] for word in my_vocab]      #Converting each word into a list so that we can use it in the multi-label binarizer/encoding
print(len(token))
print(token)
```

#### Creating a DataFrame with the encoded vocabulary

```python
encoded_df = pd.DataFrame( columns=list(my_vocab))   # Initialize an empty DataFrame with the classes as columns and columns are the unique words in the vocabulary
for sentence in sentiment_data['text']:
    words = sentence.lower().split()
    # print(f"Current sentence's words: {words}")
    
    # Create a new row with the tokenized words
    new_row = pd.Series([1 if word[0] in words else 0 for word in token], index=encoded_df.columns )
    # print(f"New row: {new_row}")
    encoded_df = pd.concat([encoded_df, new_row.to_frame().T], ignore_index=True)

    
```

# Disadvantage

```python
# it creates a spares matrix which is not very useful and can be mameory inefficient for large datasets.
# it can also lead to overfitting in some cases, especially if the vocabulary is large and the dataset is small.
# if i m purely using OHE then the encoding for each sentence will have a different length, which is not suitable for many machine learning algorithms that require fixed-length input.   so i kind of doing Gag Of Words things where i create a sparse matrix acc to the words which are present in the sentence as 1 and other as 0.
# no sematic meaning is captured in this encoding, which can lead to poor performance on tasks that require understanding of the context or meaning of the words. ex lion kill man means same as man kill lion 
# it can't deal with unseen words, which can be a problem in real-world applications where new words are constantly being introduced.
```
